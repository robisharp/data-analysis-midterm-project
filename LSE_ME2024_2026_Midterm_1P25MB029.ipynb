## clean_item_price() function

created the following function to clean the item price

```python
def clean_item_price(item_price: str):
        
    if 'each est.' in item_price:
        item_price = item_price.replace('each est.', '')


    if type(item_price) != str:
        pass
    elif '£' in item_price:
        item_price = item_price.replace('£', '')
    elif 'p' in item_price:
        item_price = item_price.replace('p', '')
        item_price = '0.'+ item_price

    # Treat cases with hyphen
    if '-' in item_price:
        #TODO: This implies I also must change the 'size' column
        item_price = item_price.split('-')[0]

    return float(item_price)
```

Save the cleaned column to a new column item-price-fixed, for need the original data to clean up the size column

### Price Distribution: Everyday Value vs Branded categories
#### Density plot shows how prices are spread across categories.

```python
tooltip_setup = (
    layer_tooltips()
    .line('@category')
    .line('[Q1 -- @median -- Q3]')
    .format('@Q1', '£ {.2f}')
    .format('@median', '£ {.2f}')
    .format('@Q3', '£ {.2f}')
)

g = (
    ggplot(
        plot_df,
        aes(
            y='category',
            x='median',
            xmin='Q1',
            xmax='Q3',
            fill='category'
        )
    ) +

    # GEOMS
    # Add a line range that 'listens to' columns informed in 'xmin' and 'xmax' aesthetics
    geom_linerange(
        size=1,
        alpha=0.75,
        tooltips=tooltip_setup
    ) +

    # Add points to the plot (listen to 'x' and 'y' and fill aesthetics)
    geom_point(
        size=4,
        stroke=0.8,
        shape=22,
        tooltips=tooltip_setup
    ) +

    # SCALES
    # Remove the legend (we can already read the categories from the y-axis)
    scale_fill_discrete(guide='none') +

    # Specify names for the axes
    scale_y_continuous(
        name='Category',
        expand=[0.05, 0.05]
    ) +
    scale_x_continuous(
        name='Price (£)',
        expand=[0., 0.05],
        format='£ {.2f}',
        breaks=np.arange(0, 20, 2.5)
    ) +

    # LABELS
    labs(
        title='Branded Products Have a Higher Median Price Than Everyday Value',
        subtitle='Dots represent the median price; bars represent the 25th and 75th percentiles'
    ) +

    theme(
        axis_text_x=element_text(size=15),
        axis_text_y=element_text(size=17),
        axis_title_x=element_text(size=21),
        axis_title_y=element_text(size=21),
        plot_title=element_text(size=20, face='bold'),
        plot_subtitle=element_text(size=21),
        legend_position='none'
    ) +

    ggsize(900, 500)
)
ggsave(g, "price_distribution.png")

g
```

### Using Boxplot to Visualize the data

```python
# Tooltip setup
tooltip_setup = (
    layer_tooltips()
    .line('@category')
    .line('Median: @median')
    .format('@median', '£ {.2f}')
)

g = (
    ggplot(
        plot_df,
        aes(
            x='category',
            y='median',
            ymin = 'Q1',
            ymax = 'Q3',
            fill='category'
        )
    ) +

    # Boxplot
    geom_boxplot(
        width=0.6,
        alpha=0.75,
        outlier_shape=21,
        outlier_size=2
    ) +

    # Remove legend
    scale_fill_discrete(guide='none') +

    # Axes
    scale_x_discrete(
        name='Category'
    ) +
    scale_y_continuous(
        name='Price (£)',
        format='£ {.2f}',
        expand=[0.05, 0.05]
    ) +

    # Labels
    labs(
        title='Price Distribution: Branded vs Everyday Value Products',
        subtitle='Boxplots show median, interquartile range, and price dispersion'
    ) +

    # Theme (kept consistent with your original)
    theme(
        axis_text_x=element_text(size=16),
        axis_text_y=element_text(size=15),
        axis_title_x=element_text(size=20),
        axis_title_y=element_text(size=20),
        plot_title=element_text(size=19, face='bold'),
        plot_subtitle=element_text(size=18),
        legend_position='none'
    ) +

    ggsize(800, 500)
)
ggsave(g, "price_distribution.png")
g
```

“The boxplot shows that Branded products not only have a higher median price, but also a wider price distribution, indicating greater variability.
Everyday Value products are priced more consistently within a narrower range.”

### Create a column to highlight Everyday Value vs Others

```python
plot_df['group'] = plot_df['category'].apply(
    lambda x: 'Everyday Value' if x == 'Everyday Value' else 'Other Categories'
)
```

We want to compare the Supermarket's own brand Everyday value against all other categories. So, creating a group column allows aclean comparison in plots. 

```python
g1 = (
    ggplot(
        plot_df,
        aes(
            x='group',
            y='median',
            fill='group'
        )
    )
    + geom_boxplot(alpha=0.75)
    + scale_fill_discrete(guide='none')
    + scale_y_continuous(
        name='Median Price (£)',
        format='£ {.2f}'
    )
    + scale_x_discrete(name='')
    + labs(
        title='Median Price Distribution: Everyday Value vs Other Categories',
        subtitle='Everyday Value products are consistently cheaper'
    )
    + theme_minimal()
    + ggsize(700, 500)
)

ggsave(g1, "price_distribution.png")

display(g1)
```

This boxplot reinforces the earlier findings by showing that Everyday Value products are priced significantly lower than products in other categories.
The tighter distribution also suggests a consistent low-price strategy, whereas other categories show greater price variation.
